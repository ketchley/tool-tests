## Part 1: What is Sentiment Analysis?
Sentiment analysis is a process that attempts to classify the overall sentiment of a piece of text as positive, negative, or neutral. There are various ways to do this, but one of the most common is to use a lexicon (also called a valence dictionary), which is essentially a long list of words that are assigned a "score" to describe how positive or negative that word is. We can then compare the words in our text(s) to the words in the lexicon and use their associated scores to get an idea of the overall sentiment of our document.

## Part 2: Sentiment Analysis Using AFINN
AFINN is one of these lexicons, and is both popular and fairly simple. It is a list of over 3300 words, each assigned a score of -5 (most negative) to 5 (most positive), and was created from 2009 to 2011 by Finn Ã…rup Nielsen. In the code blocks below, we'll explore how to use it.


```python
# this is a comment - a line (or lines) of text in a code block that won't interfere with the code

# the lines below are importing certain software packages we will need
# commands that start with ! are like running that command in a terminal (if you don't know what this means, ignore it)

!pip install pandas
import pandas as pd # pandas makes it easier to work with lots of data, create nice tables, etc.

!pip install matplotlib # a library that lets us make graphs
import matplotlib.pyplot as plt

!pip install seaborn # builds off of the previous library to make the graphs look nicer
import seaborn as sns

import os # a library to work with files and folders
from math import ceil # a specific function for later

!pip install afinn
from afinn import Afinn # afinn is the first lexicon we'll test out
afn = Afinn()
```

After running the code above, you should have something that ends with "successfully installed afinn-0.1" or something similar - at the very least, you should not have a red area show up (because that means something went wrong). If you do have that red area show up, let the TA know - otherwise, continue on!

This next code block is where the interesting things begin:

```python
# list of sentences to analyze
sentences = ['les gens pensent aux chiens', 'i hate flowers', 'he is kind and smart', 'we are kind to good people']

# go through the list and assign each sentence a score (how is it doing that?)
scores = [afn.score(sentence) for sentence in sentences]

# assign each sentence the label 'positive' if the score is above 1, 'negative' if below -1, or neutral in all other cases
# note the threshold of +/- 1 - how 
labels = ['positive' if score > 1 else 'negative' if score < -1 else 'neutral' for score in scores]

# here we are just combining the scores and sentiments we calculated above to make a nice table
table_data = list(zip(sentences, scores, labels)) # combining all three lists
afinn_intro_df = pd.DataFrame(table_data, columns=['Sentence', 'Score', 'Label']) # making the table with column titles
afinn_intro_df # displaying the table
```

In the table above, we can see the basic principle of what's happening - our code is running through a list of sentences we give it, assigning each one a score, and then labeling it as positive, negative, or neutral based on that score (and whether that score is above 1, below -1, or in the middle).

One question this might raise, though, is how those score are being determined. For example, take a look at the last two lines of the table. Why does 'hes kind and smart' get a score of 3.0, while 'we are kind to good people' get a score of 5.0? We might assume that words like 'we', 'and', 'people', etc. have no sentiment, and that the words affecting the scores would be 'kind' and 'smart' for the first sentence, and 'kind' and 'good' for the second - but why the difference in scores? Is 'good' really 2 points more positive than 'smart'? Who decided that, and how? What about the difference between sentences 1 and 2 - why is the sentence with two fairly positive words ('kind' and 'smart') as positive as the sentence with one negative word ('hate') is negative (+3 vs -3)? Is that an accurate assessment of the sentences' sentiment? Could there be alternatives?

### Running AFINN.score on each document

To explore these ideas more, let's test AFINN with a more robust dataset and longer documents. First, we need to actually find the documents we want to look at, which is what the code below does. (There won't be any output from this code block, but you still need to run it!)

```python
directory = 'watergate-test-dataset' # the name of the folder containing our documents

documents = [] # create an empty list to track documents with
for doc in os.scandir(directory): # goes into the folder with the name we gave above, and go through each filename it finds
    if not doc.name.startswith('.'): # ignores hidden files like '.ipynb_checkpoints' that are created by JupyterHub
        documents.append(doc.name) # adds that name to the list
```

```python
scores = [] # an empty list to track the sentiment scores of each document

for doc in documents: # go through each document in the list we made in the last code block
    filename = str(directory + '/' + doc) # creates a 'path' that tells the code which folder and file to look at
    with open(filename) as f: # opens that document
        contents = f.read() # gets all of the text from the file and brings it over so we can work with it
        score = afn.score(contents) # scores the file's text using AFINN's lexicon
        scores.append(score) # adds that score to the list so we can make it a column in the table

# same labeling process as we did with our first experiment with AFINN - note the thresholds for neutrality! Why might we have changed them?
labels = ['positive' if score > 3 else 'negative' if score < -3 else 'neutral' for score in scores]

# creating the table: 
table_data = list(zip(documents, scores, labels)) # combining all three lists
afinn_total_df = pd.DataFrame(table_data, columns=['File', 'Score', 'Label']) # making the table with column titles
afinn_total_df # displaying the table
```

### The Investigation Begins

As mentioned above, after getting the weirdly high and low scores from AFINN, I set out to try and determine why the scores were different and how Gale was calculating their scores. The professor and I agreed that we should include my code and notes from this process because we felt it was a good demonstration of all the small decisions that can go into a process like Sentiment Analysis and the large effects those small changes can have, as well as a good demonstration of a lot of the work that goes into digital humanities behind the scenes.

My first thought, when comparing the scores from Gale and the scores I calculated, was that perhaps the difference was due to averaging. There were two possible options I could see as being the way Gale would average the scores - by section, or by scored tokens. However, because I couldn't see any real way to differentiate tokens with a score of 0 (neutral tokens) from unscored tokens. So instead, I decided to test if the averaging system was based on document sections.

When you look at a single document in Gale's Sentiment Analysis tool, you'll see that it has a visual of the document being broken down into 20 sections, with each of those sections getting a score and being labeled as positive, negative, neutral, or not scored. I figured they might be getting the total of scored sections and dividing by the number of sections (20), so that's what I attempt to do in the code below:

```python
# averaging scores over 20 sections

scores = [] # new, empty list to track the scores
n = 20 # number of token chunks to split each document into - we use 20 because that's what Gale does

for doc in documents: # go through each document in the list (same list as last time because our documents haven't changed)
    filename = str(directory + '/' + doc)  # creates a 'path' that tells the code which folder and file to look at
    with open(filename) as f: # lets us access the actual text inside the file
        contents = f.read() # gets all of the text from the file and brings it over so we can work with it
        
        bag_o_words = contents.split() # separate into "words" (actually tokens) wherever there is a space
        size = ceil(len(bag_o_words) / n) # determining the size of each chunk, aka how many tokens to include in it
        chunk_words = list(map(lambda x: bag_o_words[x * size:x * size + size], list(range(n)))) # splitting into chunks
        
        formatted_chunks = [] # list to created formatted chunks in, b/c chunks above are just a list of words, not one string
        for chunk in chunk_words:
            combined = ' '.join([str(word) for word in chunk]) # combine the words in the list with a space separating them
            formatted_chunks.append(combined) # add that formatted chunk to our new list
            
        sub_scores = [afn.score(chunk) for chunk in formatted_chunks] # score the chunk
        total_score = sum(sub_scores) # sums all the chunks' scores - is actually just the normal score from above
        avg_score = total_score/20 # average the score by dividing by the number of chunks (20)
        scores.append(avg_score) # adds that score to the list so we can make it a column in the table

# same labeling process as we did with our first experiment with afinn - note the threshold though!
labels = ['positive' if score > 0.1 else 'negative' if score < -0.1 else 'neutral' for score in scores]

# creating a table: 
table_data = list(zip(documents, scores, labels)) # combining all three lists
afinn_avg_df = pd.DataFrame(table_data, columns=['File', 'Average Score', 'Label']) # making the table with column titles
afinn_avg_df # displaying the table
```

That initially looks a lot more correct - there's no immediate scores way above or below what Gale can give us, which is a good sign. Since pandas truncates the table, though, let's check and see if that assumption holds true by graphing the distribution.

```python
sns.set(font_scale=1.3) # increasing the text size so it's easier to read

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10)) # sizes the graphs nicely and creating the two spaces to graph in
sns.swarmplot(ax=ax1, x="Average Score", y="Label", hue="Label", palette="magma", data=afinn_avg_df) # making the swarmplot
# a swarmplot is a categorical scatterplot with points adjusted to not overlap so it's easier to see the distribution
ax1.set_title('AFINN Scores Averaged Over 20 Chunks') # adding a title to the graph

sns.boxplot(ax=ax2, x=afinn_avg_df["Average Score"]) # making the boxplot
print() # printing blank line to remove weird '<Axes.subplot> text
```

Well, it looks like we were close, but not quite right - even though the scores are much closer to what they should be (according to Gale at least), there are still values that fall outside of Gale's range of -5 to 5 (for example, the two documents on the far left that have scores of less than -6). 

That's not great, because it means we can't just average the scores for 20 equal sections of our document. But what should we do?

We decided the next best step was to narrow down to just one document, and to look specifically at what was happening in it - maybe the words/tokens weren't being scored correctly for some reason?

From our searches, we can see that we originally gave this document a score of -12, so the sum of all tokens in this document come out to -12, at least for us. We can also see that the average-over-20-sections version scores it as -0.6 -  closer to 0, but definitely not quite right.

Next, we moved on to getting the scores of each token in the document:

```python
# looking at scores of tokens

document = 'Calvi_allegations_in_CS101944136.txt' # choosing a document to look at

# getting sentiment scores for the text of each document
words = [] # to track the words
scores = [] # to track the scores of those words
num_words = 0 # the number of tokens in the document

filename = str(directory + '/' + document) # creating the filepath to use
with open(filename) as f: # lets us access the actual text inside the file
    contents = f.read() # gets all of the text from the file and brings it over so we can work with it

    bag_o_words = contents.split() # dividing the text into "words" (tokens, really, as we're splitting wherever there is whitespace)
    num_words = len(bag_o_words) # finding the number of tokens for use later

    for word in bag_o_words: # go through each token
        words.append(word) # add the word to a list to use as a column later
        scores.append(afn.score(word)) # scoring the word and tracking that score
            
print("Score divided by total words", round(total_score/num_words, 2))
print() 

table_data = list(zip(words, scores)) # combining all three lists
doc_words_df = pd.DataFrame(table_data, columns=['"Word"', 'AFINN Score']) # making the table with column titles
doc_words_df # displaying the table
```

Yep, from the graph above, there are no words getting scores outside of the range of -3 to 3, which is within the correct range for AFINN. Which means that the difference in scores is not an issue with AFINN not scoring words.

So at this point, we knew that it wasn't an average of the 20 section scores, and we knew the words were being scored correctly, and we knew that dividing by the total number of tokens wasn't quite right either. The only other thing we could think of was that Gale was somehow averaging by the number of scored tokens.

When you open an individual document in Gale's Sentiment Analysis tool, it lists the total number of tokens (69 in the case of the document above, which aligns with what we found) and the number of scored tokens (which, for the document above, was 6). This was pretty confusing to me, as I couldn't understand how they were separating scored tokens from unscored tokens. After all, from the investigation directly above, looking at individual tokens for one document, each token got a score. It was possible that unscored tokens were also getting a score of 0 (afterall, 'Calvi', the first token in the document, got a score of 0, and it seems unlikely that was in AFINN's lexicon to get a score for), but if that was the case, how was Gale differentiating between unscored tokens and neutral tokens that got an actual score of 0?

### The Solution ... ?

It was at this point that we decided to reach out to Gale themselves to try and figure out what was going on - namely, how they were determing a token that was scored with a score of 0 from an unscored token. This was when we found a major source of the confusion - words cannot receive a score of 0 with AFINN. Documents with a score of 0 are neutral, but words with a score of 0 actually meant they were unscored. (This is much clearer with Gale's backend system, where unscored words get a score of NaN, or Not a Number, rather than 0). We also got confirmation from Gale that the way they were getting document score had nothing to do with the section scores, and was instead just the sum of the scores divided by the number of scored tokens.

This seemed easy enough to apply, so we decided to do just that:

```python
# dividing by scored tokens

document = 'Calvi_allegations_in_CS101944136.txt' # choosing a document to look at

# getting sentiment scores for the text of each document
words = []
scores = []
total_words = 0

filename = str(directory + '/' + document)
with open(filename) as f: # lets us access the actual text inside the file
    contents = f.read() # gets all of the text from the file and brings it over so we can work with it

    bag_o_words = contents.split() # dividing the text into "words" (tokens, really, as we're splitting wherever there is whitespace)
    total_words = len(bag_o_words)
    
    for word in bag_o_words:
        score = afn.score(word) # score the word
        if score != 0: # if the score isn't 0 (which means the token wasn't scored)
            words.append(word) # add the word to the word list
            scores.append(score) # add the score to the score list
            
total_score = sum(scores) # add all the scores together
num_scored_words = len(scores) # find how many words there are in the list (which are the scored words)
doc_score = total_score / num_scored_words # get the document score using Gale's method

#print("Score divided by scored words", scored_words)
print("Sum of all scores:", total_score)
print("Number of scored tokens:", num_scored_words)
print("Document score (total / number of scored tokens):", doc_score)
print()

# just so we can see what the individual scores are and what words were scored
table_data = list(zip(words, scores)) # combining all three lists
doc_scored_tokens_df = pd.DataFrame(table_data, columns=['"Word"', 'Doc Score']) # making the table with column titles
doc_scored_tokens_df # displaying the table
```

We found that the sum of all the scores was -12 still, and we found that the number of scored tokens was 12 (which was easy enough to find now that we knew to ignore any tokens with a score of 0).

However, we ran into yet another issue. Applying Gale's document score formula (sum of the scores / number of scored tokens) would mean computing -12 / 12 - and as I'm sure you've noticed, -12 / 12 does not equal 0. 

So even after all of this, even after we applied the same formula Gale uses, we're still getting a different score from Gale - why? As we found out, the difference comes down to the lexicons we use. While our code found 12 tokens with scores, Gale says they only have 6 scored tokens for the same document. The reason for that difference is the size of the lexicon - the Python version of AFINN, the one we're using, is based on AFINN-165, which has a lexicon containing around 3,400 terms. However, the professor happens to know that Gale still uses AFINN-111, which has around 2,400 terms. Basically, our version of AFINN can recognize and score about 1,000 more words than Gale's, which explains why our version can score 12 tokens while Gale's can only score 6. 

Essentially, the conclusion at the end of all this is that, while sentiment analysis may seem like this thing with a correct answer and a clear right or wrong, it's actually made of a lot of seemingly small choices that can cause huge differences in results and analyses. Whether it be the decision of what lexicon or what version of a lexicon to use (AFINN-111 vs 165), how to average or combine the token scores to get a document score (across sections, by total tokens, or by scored tokens), or what threshold to use for different labels (1 and -1? 3 and -3? 0.1 and -0.1?), we hope this demo and exploration have helped show why it's so important to track and understand the decisions you make in the course of a digital humanities project.

## Optional Additional Exploration
### Sentiment Analysis with VADER (Valence Aware Dictionary and sEntiment Reasoner)

VADER is another way of classifying the sentiment of texts. Unlike AFINN, VADER tells users how positive, negative, and neutral it feels the text is, rather than just the sum of all the scored tokens in the text. We explore what it looks like more in the code below.

```python
# don't worry about this code either, just run it - it installs the things we need to use later

!pip install vaderSentiment
import vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

Again, you should ideally have something similar to "successfully installed vaderSentiment" - if not, reach out to the TA.

```python
analyser = SentimentIntensityAnalyzer() # instantiating the tool

# the same list of sentences as we used for AFINN
sentences = ['les gens pensent aux chiens', 'i hate flowers', 'he is kind and smart', 'we are kind to good people']

scores = [] # empty list to track the scores we get back
for sentence in sentences: # go through each sentence
    scores.append(analyser.polarity_scores(sentence)) # track the scores for each sentence
    
"""
VADER's scoring function creates what Python calls a 'dictionary' - essentially a set of keys and values associated with
those keys. This means that we don't just get a single score - instead, we get a set of scores that includes the negative
score, positive score, neutral score, and compound (overall) score for a single document. To get those all into rows of a table
for easier viewing, we go through each dictionary in the 'scores' list and make a list of all the negative scores,
then of all the neutral scores, etc., which is the code below.
"""
neg_list = [d['neg'] for d in scores]
neu_list = [d['neu'] for d in scores]
pos_list = [d['pos'] for d in scores]
compound_list = [d['compound'] for d in scores]

# assign label based on VADER's compound score - note the threshold of 0.1 - how did we decide that? What if we change it?
labels = ['positive' if score > 0.1 else 'negative' if score < -0.1 else 'neutral' for score in compound_list]

# making and displaying the table
vader_data = list(zip(sentences, neg_list, neu_list, pos_list, compound_list, labels)) # combining all lists
vader_intro_df = pd.DataFrame(vader_data, columns=['Sentence', 'Percent Negative', 'Percent Neutral', 'Percent Positive',
                                                   'Compund Score', 'Label']) # making the table with column titles
vader_intro_df # displaying the table
```

Above, we can see the results VADER gives when looking at the same original 4 sentences we used when introducing AFINN. Just for fun, let's also look at what AFINN's table for those sentences looked like:

```python
afinn_intro_df # this is the table we made all the way back in part 2 - displaying it here for easier comparison
```

### VADER with a Larger Content Set
To get a better sense of VADER, let's also look at it with a larger content set, same as we did with AFINN. This uses the same content set we used with AFINN.

```python
analyser = SentimentIntensityAnalyzer() # instantiating the tool

scores = [] # a list to keep track of the documents
for doc in documents: # go through each document in the list we made when working with AFINN
    filename = str(directory + '/' + doc) # appending the document name to the directory name so we can find the file
    with open(filename) as f: # lets us access the actual text inside the file
        contents = f.read() # gets all of the text from the file and brings it over so we can work with it
        score_dict = analyser.polarity_scores(contents) # scores the file's text
        scores.append(score_dict) # adds that score to the list so we use the scores later
        
# same thing as above - extracting each score type into a list
neg_list = [d['neg'] for d in scores]
neu_list = [d['neu'] for d in scores]
pos_list = [d['pos'] for d in scores]
compound_list = [d['compound'] for d in scores]

# labeling the documents based on their compound score - note the threshold for what is neutral and not
labels = ['positive' if score > 0.1 else 'negative' if score < -0.1 else 'neutral' for score in compound_list]

table_data = list(zip(documents, neg_list, neu_list, pos_list, compound_list, labels)) # combining all lists
vader_df = pd.DataFrame(table_data, columns=['File', 'Percent Negative', 'Percent Neutral', 'Percent Positive',
                                             'Overall Score', 'Label']) # creating the table
vader_df # displaying the table
```

### AFINN

```python
# creating lists to track the total score, number of scored tokens, and document score for each document
total_scores = []
num_scored = []
doc_scores = []

for doc in documents: # go through each document in the list we made in the last code block
    token_scores = []
    
    filename = str(directory + '/' + doc)
    with open(filename) as f: # lets us access the actual text inside the file
        contents = f.read() # gets all of the text from the file and brings it over so we can work with it
        bag_o_words = contents.split() # dividing the text into "words" (tokens, really, as we're splitting wherever there is whitespace)

        for word in bag_o_words:
            score = afn.score(word) # score the word
            if score != 0: # if the score isn't 0 (which means the token wasn't scored)
                token_scores.append(score) # add the score to the score list
            
        total_scores.append(sum(token_scores)) # add all the scores together
        num_scored.append(len(token_scores)) # find how many words there are in the list (which are the scored words)
        if len(token_scores) == 0: # so we don't end up dividing by 0
            doc_scores.append(0)
        else: # if we scored more than 0 tokens
            doc_scores.append(round(sum(token_scores) / len(token_scores), 3)) # get the document score using Gale's method
            # we also round to three decimal points because that's what Gale does

labels = ['positive' if score > 1 else 'negative' if score < -1 else 'neutral' for score in doc_scores]            

# creating the table
table_data = list(zip(documents, total_scores, num_scored, doc_scores, labels)) # combining all lists
# making the table with column titles
student_afinn_df = pd.DataFrame(table_data, columns=['File', 'Total Score', 'Num Scored Tokens', 'Doc Score', 'Label'])
student_afinn_df # displaying the table
```

```python
# because pandas only shows the first and last 5 rows of a table
# the code below helps us get a better understanding of the distribution of sentiment labels

print("Sentiment Label Counts")
print(student_afinn_df['Label'].value_counts()) # get a list of unique values for that column and the number of times each appeared
```

```python
sns.set(font_scale=1.3) # increasing the text size so it's easier to read

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10)) # sizes the graphs nicely and creating the two spaces to graph in
sns.swarmplot(ax=ax1, x="Doc Score", y="Label", hue="Label", palette="magma", data=student_afinn_df) # making the swarmplot
# a swarmplot is a categorical scatterplot with points adjusted to not overlap so it's easier to see the distribution
ax1.set_title('AFINN Scores for Your Content Set, Default') # adding a title to the graph

sns.boxplot(ax=ax2, x=student_afinn_df["Doc Score"]) # making the boxplot
print() # printing blank line to remove weird '<Axes.subplot> text
```
